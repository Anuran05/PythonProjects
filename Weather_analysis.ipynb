```python
#Find all instances when snow was recorded 
import pandas as pd
data = pd.read_csv(r"C:\Users\user\Downloads\file.csv")

#data['Weather'].value_counts()
#data[data['Weather']=='Snow'] #Shows all values in Weather column containing precisely the word Snow 
data[data['Weather'].str.contains('Snow')] #Shows all values in Weather column containing the word snow anywhere in them
```

```python
#Mean value of each column against each Weather condition
#data['Weather']=data['Weather'].fillna(0) #Replace NaN values with zeros for a column
import pandas as pd
data = pd.read_csv(r"C:\Users\user\Downloads\file.csv")
data.groupby('Weather').mean() #it appears that this error may be associated with the most recent releases of pandas. I suggest considering a downgrade of the pandas package, such as reverting to version 1.3.5.
```
